This technology provides instant feedback on exercise form, tracks workout routines, and measures performance metrics, optimizing training sessions for users and trainers alike.

### Workouts Monitoring Code

```python
import cv2
from ultralytics import YOLO, solutions

# Load the pre-trained YOLOv8 model
model = YOLO("yolov8n-pose.pt")

# Open the video file
cap = cv2.VideoCapture("path/to/video/file.mp4")
assert cap.isOpened(), "Error reading video file"

# Get video properties: width, height, and frames per second (fps)
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# Initialize the video writer to save the output video
video_writer = cv2.VideoWriter("object_counting_output.avi", cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))

# Initialize the workouts monitoring object with angle adjustment, keypoint selection and other parameters
gym_object = solutions.AIGym(
    line_thickness=2, # Thickness of the keypoints visualization drawn
    view_img=True, # Display the image during processing
    pose_type="pushup", # Selection of pose_type for monitoring {"pushup", "pullup", "abworkout", "squat"}
    kpts_to_check=[6, 8, 10], # Selection of keypoints for workout monitoring.
)

# Process video frames in a loop
while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video frame is empty or video processing has been successfully completed.")
        break

    # Perform object tracking on the current frame
    tracks = model.track(im0, show=False)

    # Use the workouts out monitoring object to monitor workouts in the frame
    im0 = gym_object.start_counting(im0, tracks)

    # Write the annotated frame to the output video
    video_writer.write(im0)

# Release the video capture and writer objects
cap.release()
video_writer.release()

# Close all OpenCV windows
cv2.destroyAllWindows()
```
