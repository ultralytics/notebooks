# Object Detection with KITTI Dataset using Ultralytics YOLO11

This notebook serves as an initial step for training the YOLO11 model on the [KITTI](https://docs.ultralytics.com/datasets/detect/construction-ppe/) detection dataset. KITTI dataset is one of the most influential benchmark datasets for autonomous driving and computer vision. Released by the Karlsruhe Institute of Technology and Toyota Technological Institute at Chicago, it contains stereo camera, LiDAR, and GPS/IMU data collected from real-world driving scenarios.

## Dataset Structure

The KITTI dataset dataset is divided into two main subsets:

- **Training set**: Consisting of 5985 images, each accompanied by corresponding annotations.
- **Validation set**: Comprising 143 images, with annotations paired for each one.

## Applications

The KITTI dataset drives advancements in autonomous driving and robotics by enabling key computer vision tasks. It supports vehicle perception for detecting and tracking objects, enhances 3D scene understanding through depth estimation and localization, and improves motion prediction for dynamic environments. Additionally, it serves as a benchmark for evaluating performance in tasks like object detection and tracking, making it a cornerstone for computer vision research.

## Train

Train YOLO11 on [Detect](https://docs.ultralytics.com/tasks/detect/), [Segment](https://docs.ultralytics.com/tasks/segment/), [Classify](https://docs.ultralytics.com/tasks/classify/) and [Pose](https://docs.ultralytics.com/tasks/pose/) datasets. See [YOLO11 Train Docs](https://docs.ultralytics.com/modes/train/) for more information.

````
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data="kitti.yaml", epochs=10, imgsz=640)
````

![KITTI dataset sample image](https://github.com/ultralytics/docs/releases/download/0/kitti-dataset-sample.avif)

## Predict

YOLO11 can be used from the command line with the `yolo` CLI command for various tasks. It supports additional arguments (e.g., imgsz=640). See a full list of available [yolo arguments](https://docs.ultralytics.com/usage/cfg/) and other details in the [YOLO11 Predict Docs](https://docs.ultralytics.com/modes/train/).

````
from ultralytics import YOLO

# Load a model
modelp = YOLO(f"{model.trainer.save_dir}/weights/best.pt")  # load a fine-tuned model

# Inference using the model (img)
prediction_results = modelp.predict("https://ultralytics.com/assets/kitti-inference-im0.png", save=True)

# Inference using the model (video)
# prediction_results = modelp.predict("https://ultralytics.com/assets/kitti-inference-vid.mp4", save=True)
````
